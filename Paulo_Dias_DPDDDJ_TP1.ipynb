### Ponto de partida

Neste teste, é apresentado um estudo de caso que precisa ser corrigido.
O sistema de gerência de uma biblioteca está mal projetado, apresentando alto acoplamento, repetição de código e dificuldades de manutenção.

O sistema de biblioteca possui as seguintes funcionalidades:

* Gerenciamento de livros (adicionar, remover, buscar);
* Empréstimo de livros para usuários;
* Rastreamento de empréstimos vencidos;
* Gerenciamento de usuários (adicionar, remover, atualizar).

Problemas Identificados no Diagrama Atual:

1. Alto Acoplamento: Classes estão diretamente interligadas, fazendo referências diretas umas às outras para realizar suas funções;
2. Repetição de Código: Lógica de busca e remoção de livros e usuários é replicada em várias partes do sistema;
3. Dificuldades de Manutenção: Qualquer mudança em uma classe, como adicionar um novo tipo de item para empréstimo (e.g., DVDs), requer modificações substanciais em várias outras classes.

#### Justificativas

Factory Method – Utilizado para centralizar a instanciação de objetos, proporcionando melhor controle e acesso a eles a partir de um único ponto. No caso da LibraryItemFactory, foi especialmente útil, pois a criação de novas subfactories, como BookFactory e DVDFactory (já presentes no código), permite adicionar novos tipos de itens à biblioteca sem a necessidade de modificar grandes partes do sistema. Basta criar uma nova fábrica para o subproduto e associá-la ao seu respectivo modelo, garantindo também seu correto armazenamento.<br>

Singleton – Garante que apenas uma única instância da classe Library exista durante toda a execução do sistema, evitando a criação indevida de múltiplas instâncias que poderiam comprometer a integridade dos dados.<br>

Facade – Simplifica e controla o acesso à classe principal Library. A classe LibraryFacade expõe apenas os métodos necessários, evitando que classes externas interajam diretamente com a implementação interna da biblioteca.<br>

Observer – Implementa um mecanismo de assinatura para notificações, permitindo que usuários interessados (subscritos) recebam atualizações sobre novidades. Essa funcionalidade é implementada através de uma interface que garante que apenas os usuários registrados recebam as notificações.<br>

Strategy – Centraliza e abstrai os algoritmos de busca, tornando a implementação mais flexível e facilitando a integração com o front-end. Isso melhora a legibilidade, a manutenibilidade e a extensibilidade do código.
